```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
```

```python
def predict(symbol):
  df_f=df[['Date', symbol]].copy()
  df_f.columns=["ds", "y"]

  prophet=Prophet(daily_seasonality=True)
  prophet.fit(df_f)

  future_dates=prophet.make_future_dataframe(periods=30)
  predictions=prophet.predict(future_dates)
  predictions_filtered=predictions.iloc[-28:]

  closing_prices=market_prices[['Date', symbol]].copy()
  closing_prices.columns=["ds", "Closing_prices"]

  merged_df=pd.merge(predictions_filtered, closing_prices, on='ds', how='left')
  merged_df=merged_df.dropna()
  merged_df = merged_df[['ds', 'yhat', 'yhat_lower', 'yhat_upper', 'Closing_prices']]

  print(f'\nPredictions and market closing prices comparison  for the company {symbol}: \n')
  print(merged_df)

  y_true = merged_df['Closing_prices']
  y_pred = merged_df['yhat']

  mae = mean_absolute_error(y_true, y_pred)
  mse = mean_squared_error(y_true, y_pred)
  rmse = np.sqrt(mse)
  mape = np.mean(np.abs((y_true - y_pred) / y_true)) * 100

  print(f'\nModel Evaluation Metrics for {symbol}:')
  print(f'MAE: {mae:.2f}')
  print(f'RMSE: {rmse:.2f}')
  print(f'MAPE: {mape:.2f}%')
```

```python
for symbol in symbols:
  predict(symbol)
```
